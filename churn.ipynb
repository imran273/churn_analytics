```python
import pandas as pd
import numpy as np
churn_df = pd.read_csv('/home/subham/PycharmProjects/churn_analytics/WA_Fn-UseC_-Telco-Customer-Churn.csv',
                       )
col_names = churn_df.columns.tolist()

print ("Column names:")
print (col_names)


print ("\nSample data:")
churn_df.head()
```

```python
# churn_df.loc[churn_df['Churn'] == 'Yes' , ['Churn']] = 1
# churn_df.loc[churn_df['Churn'] == 'No' , ['Churn']] = 0

y= churn_df['Churn']
print(y.head())

to_drop = ['customerID','TotalCharges','PaperlessBilling','Churn']
churn_feat_space = churn_df.drop(to_drop,axis=1)

yes_no_cols = ['Partner', 'Dependents','PhoneService', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
               ]
churn_feat_space[yes_no_cols] = churn_feat_space[yes_no_cols] == 'Yes'


churn_feat_space.loc[churn_feat_space['PaymentMethod'] == 'Electronic check' , 'PaymentMethod'] = 0
churn_feat_space.loc[churn_feat_space['PaymentMethod'] == 'Mailed check' , 'PaymentMethod'] = 0
churn_feat_space.loc[churn_feat_space['PaymentMethod'] == 'Bank transfer (automatic)' , 'PaymentMethod'] = 1
churn_feat_space.loc[churn_feat_space['PaymentMethod'] == 'Credit card (automatic)' , 'PaymentMethod'] = 1


churn_feat_space.loc[churn_feat_space['InternetService'] == 'Fiber optic' , 'InternetService'] = 2
churn_feat_space.loc[churn_feat_space['InternetService'] == 'DSL' , 'InternetService'] = 1
churn_feat_space.loc[churn_feat_space['InternetService'] == 'No' , 'InternetService'] = 0

churn_feat_space.loc[churn_feat_space['Contract'] == 'Month-to-month' , 'Contract'] = 0
churn_feat_space.loc[churn_feat_space['Contract'] == 'One year' , 'Contract'] = 1
churn_feat_space.loc[churn_feat_space['Contract'] == 'Two year' , 'Contract'] = 2

churn_feat_space.loc[churn_feat_space['gender'] == 'Female' , 'gender'] = 0
churn_feat_space.loc[churn_feat_space['gender'] == 'Male' , 'gender'] = 1


# churn_feat_space.drop('Churn',axis=1)

print(churn_feat_space.head())
```

```python
features = churn_feat_space.columns
X = churn_feat_space.as_matrix().astype(np.float)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)

print ("Feature space holds %d observations and %d features" % X.shape)
print ("Unique target labels:", np.unique(y))
```

```python
from sklearn.cross_validation import KFold

def run_cv(X,y,clf_class,**kwargs):
    # Construct a kfolds object
    kf = KFold(len(y),n_folds=5,shuffle=True)

    # Iterate through folds
    for train_index, test_index in kf:
        X_train, X_test = X[train_index], X[test_index]
        y_train,y_test = y[train_index],y[test_index]
        # Initialize a classifier with key word arguments
        clf = clf_class(**kwargs)
        clf.fit(X_train,y_train)
        y_pred = clf.predict(X_test)
    return y_test,y_pred
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier as RF
from sklearn.neighbors import KNeighborsClassifier as KNN

def accuracy(y_true,y_pred):
    # NumPy interprets True and False as 1. and 0.
    return np.mean(y_true == y_pred)

print ("Support vector machines:")
y_true,y_pred=run_cv(X,y,SVC)
print ("%.3f" % accuracy(y_true,y_pred))
print ("Random forest:")
y_true,y_pred=run_cv(X,y,RF)
print ("%.3f" % accuracy(y_true,y_pred))
print ("K-nearest-neighbors:")
y_true,y_pred=run_cv(X,y,KNN)
print ("%.3f" % accuracy(y_true,y_pred))
```
